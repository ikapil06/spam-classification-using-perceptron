```python
#Assigning values to x and y to train_test split
x=df.drop(labels="1",axis=1)
y=df['1']
```

```python
# train_test splitting
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.30,random_state=101)
```

```python
#Importing perceptron
from sklearn.linear_model import Perceptron

clf = Perceptron(eta0=1.0,max_iter=1000)
clf.fit(x_train, y_train)
```

```python
pred_p=clf.predict(x_test)
print(pred_p)
```

```python
#Evaluate a score by cross-validation.
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold

# define model evaluation method
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
# evaluate model
scores = cross_val_score(clf, x, y, scoring='accuracy', cv=cv, n_jobs=-1)

```

```python
from numpy import mean
from numpy import std
# summarize result
print('Mean Accuracy: %.4f (%.4f)' % (mean(scores), std(scores)))
```

## Hyperparameter tuning

```python
# defining the grid
grid = dict()
grid['eta0'] = [0.0001, 0.001, 0.01, 0.1, 1.0]
```

```python
from sklearn.model_selection import GridSearchCV

search = GridSearchCV(clf, grid, scoring='accuracy', cv=cv, n_jobs=-1)
results = search.fit(x, y)
print('Mean Accuracy: %.3f' % results.best_score_)
print('Config: %s' % results.best_params_)
means = results.cv_results_['mean_test_score']
params = results.cv_results_['params']
for mean, param in zip(means, params):
    print(">%.3f with: %r" % (mean, param))
```

### Hyperparameter on Epochs 

```python
grid = dict()
grid['max_iter'] = [1, 10, 100, 1000, 10000]
```

```python
search = GridSearchCV(clf, grid, scoring='accuracy', cv=cv, n_jobs=-1)
results = search.fit(x, y)
print('Mean Accuracy: %.3f' % results.best_score_)
print('Config: %s' % results.best_params_)
means = results.cv_results_['mean_test_score']
params = results.cv_results_['params']
for mean, param in zip(means, params):
    print(">%.3f with: %r" % (mean, param))
```
